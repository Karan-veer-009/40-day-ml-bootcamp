# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
# Creating a simple figure by giving the x and y co-ordinates
fig, ax =plt.subplots()
ax.plot([1,2,3,4],[1,2,5,4])
plt.show()
```

```python
#Figures
fig,axs=plt.subplot_mosaic([['left','right_top'],
                           ['left','right_bottom']])
```

```python
#Axes
ax.set_title("Simple Plot")
ax.set_xlabel("x")
ax.set_ylabel("y")
import matplotlib.ticker as ticker
#Axis- Formatter+Locator
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(1))
ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%.1f'))
```

```python
# Plotting functions
b=np.matrix([[23,34],[45,34]])
b_asarray = np.asarray(b)

np.random.seed(19680801)  # seed the random number generator.
data = {'a': np.arange(50), # Sequence number till 50
        'c': np.random.randint(0, 50, 50),# 50 random numbers bw 0 and 50
        'd': np.random.randn(50)} # 50 random number  from std
data['b'] = data['a'] + 10 * np.random.randn(50) # adding some noise to make scattered line-like pattern
data['d'] = np.abs(data['d']) * 100 # making it postive and scaling it up by 100

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
# width=5, height=2.7
ax.scatter('a', 'b', c='c', s='d', data=data)
# colour =c,size=d,x=a,y=b
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

```python
# Two approaches
#oo-Style
x = np.linspace(0, 2, 100)  # Sample data.

# Note that even in the OO-style, we use `.pyplot.figure` to create the Figure.
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x, label='linear')  # Plot some data on the Axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the Axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the Axes.
ax.set_ylabel('y label')  # Add a y-label to the Axes.
ax.set_title("Simple Plot")  # Add a title to the Axes.
ax.legend()  # Add a legend.
```

```python
#pyplot styple
x = np.linspace(0, 2, 100)  # Sample data.

plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x, label='linear')  # Plot some data on the (implicit) Axes.
plt.plot(x, x**2, label='quadratic')  # etc.
plt.plot(x, x**3, label='cubic')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
def my_plotter(ax, data1, data2, param_dict):
    out =ax.plot(data1,data2,**param_dict)
    return out

data1, data2, data3, data4 = np.random.randn(4, 100)  # make 4 random data sets
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))
my_plotter(ax1, data1, data2, {'marker': 'x'})
my_plotter(ax2, data3, data4, {'marker': 'o'})
ax1.text(0,1,r'$\mu=115$')
```

```python
#Annotations
fig, ax = plt.subplots(figsize=(5, 2.7))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t, s, lw=2,label='data1')

ax.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_ylim(-2, 2)
#mathematical expressions in text
ax.set_title(r'$\sigma_i=15$')
ax.set_yscale('log')
ax.legend()
```
