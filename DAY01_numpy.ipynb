# Mini Project

```python
import numpy as np
class Preprocessor:
  def __init__(self,data):
    self.data=np.array(data,dtype=float)
  def summary(self):
    return {"mean":self.data.mean(axis=0).tolist(),#Converting the mean of columns to a list with th help of tolist()
    "std":self.data.std(axis=0).tolist(),
    "min":self.data.min(axis=0).tolist(),
    "max":self.data.max(axis=0).tolist()}
  def normalize(self):
    return {"normalize":((self.data-(self.data.min(axis=0)))/(self.data.std(axis=0)+1e-8)).tolist()}
  def detect_outliers(self,z_thresh=3):
    # The 1e-8 is added to avoid being divided by zero
    z=np.abs(self.data-self.data.mean(axis=0))/(self.data.std(axis=0)+1e-8)
    return np.where(z>z_thresh)

data =np.array([
    [5,20,100],
    [9,78,2345],
    [0,98,908]
])
output=Preprocessor(data)
print(output.detect_outliers())
```
